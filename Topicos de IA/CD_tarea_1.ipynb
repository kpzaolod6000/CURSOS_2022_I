## Distancia Manhattan

```python
def distManhattan(x1,x2):
  sum = 0
  len_ = len(x1)
  for i in range(len_):
    if(not pd.isnull(x1[i]) and not pd.isnull(x2[i])):
      sum += abs(x1[i] - x2[i])
  return sum
```

## Distancia Euclideana

```python
def distEuclideana(x1,x2):
  sum = 0
  len_ = len(x1)
  for i in range(len_):
    if(not pd.isnull(x1[i]) and not pd.isnull(x2[i])):
        sum += (x1[i] - x2[i])**2
  return math.sqrt(sum)
```

```python
def AproxPearson(x,y):
  n_Reg = len(x)
  
  x_y = 0 # x*y
  s_x = 0 # sumatoria(x)
  s_y = 0 # sumatoria(y)
  s_x2 = 0 ## sumatoria de cuadrados(x)
  s_y2 = 0 ## sumatoria de cuadrados(x)
  n = 0
  for i in range(n_Reg):
    if(not pd.isnull(x[i]) and not pd.isnull(y[i])):
      n += 1
      x_y += (x[i] *  y[i])
      s_x += x[i]
      s_y += y[i]
      s_x2 += x[i]**2
      s_y2 += y[i]**2

      # print(x[i],' * ', y[i] ,' : ',x_y)
        

  s_x_s_y = (s_x * s_y)/n # sumatoria(x*y)/n
  s_x2N = (s_x**2)/n # sumatoria de X al cuadrado sobre n
  s_y2N = (s_y**2)/n # sumatoria de Y al cuadrado sobre n
  
  # print(s_x_s_y)
  # print(s_x2N)
  # print(s_y2N)
  r = (x_y - s_x_s_y) / (math.sqrt(s_x2 - s_x2N) * math.sqrt(s_y2 - s_y2N))
  return r
# AproxPearson('Angelica','Jordyn')
```

## Similitud del Coseno

```python
def similCoseno(x,y):
  n_Reg = len(x)
  x_y = 0
  x2 = 0
  y2 = 0
  for i in range(n_Reg):
      x_y += x[i] * y[i]
      x2 += x[i]**2
      y2 += y[i]**2

  # print(x_y)
  # print(math.sqrt(x2))
  # print(math.sqrt(y2))
  cos_ = (x_y)/(math.sqrt(x2) * math.sqrt(y2))
  return cos_
```

```python
def cosBetweenPersons(person1,person2):
  x1 = data[person1]
  x2 = data[person2]
  x = x1.to_numpy()
  y = x2.to_numpy()
  x_new = [0 if pd.isnull(i) else i for i in x]
  y_new = [0 if pd.isnull(i) else i for i in y]

  
  return similCoseno(x_new,y_new)
```
